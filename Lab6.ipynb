```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
!rm -rf /content/result
inputFile= "/content/ReviewsSample.csv"
outputPath = "result"
lines = sc.textFile(inputFile)
#lines.collect()
```

```python
#Transposes the original Amazon food dataset, obtaining an RDD of pairs (tuples) of the type:(user_id, list of the product_ids reviewed by user_id)
header = lines.first()
data = lines.filter(lambda row: row != header)
newdataset = data.map(lambda line: (line.split(',')[2], line.split(',')[1]) )
newdataset = newdataset.distinct()
newdataset2 = newdataset.groupByKey()
newdataset2 = newdataset2.mapValues(lambda x: list(x))
newdataset2.saveAsTextFile(outputPath)
```

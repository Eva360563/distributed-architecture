```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
inputFile= "/content/part-r-00000"
lines = sc.textFile(inputFile)
```

```python
lines.collect()
```

```python
ho_lines= lines.filter(lambda e: e.startswith('ho'))
```

```python
ho_lines.collect()
```

```python
#The number of selected lines
numb= ho_lines.count()
numb
```

```python
#The maximum frequency (maxfreq) among the ones of the selected lines (i.e., the maximum value of freq in the lines obtained by applying the filter).
maxfreq = ho_lines.map(lambda line: int(line.split('\t')[1])).max()
maxfreq
#map mi restituisce tutte le frequenze e max la più alta
```

```python
greater_freq = ho_lines.filter(lambda line: int(line.split('\t')[1]) > 0.8*maxfreq)
greater_freq.collect()
```

```python
num_lines= greater_freq.count()
num_lines
```

```python
gr = greater_freq.map(lambda line: line.split('\t')[0]).collect()
gr
```

```python
from pyspark import SparkContext, SparkConf

conf = SparkConf().setAppName("Application name")
sc = SparkContext.getOrCreate(conf=conf)

#es 30
inputFile = sc.textFile("log.txt")
inputFile.collect()
```

```python
filteredLine = inputFile.filter(lambda s: s.find ('google')>=0)
filteredLine.collect()
```

```python
outputpath = "output.txt"
filteredLine.saveAsTextFile(outputpath)
```

```python
#es 31
inputFile = sc.textFile("log.txt")
inputFile.collect()
```

```python
filterRDD = inputFile.filter(lambda s: s.find('www.google.com')>=0)
onlyIP = filterRDD.map(lambda x: x.split('-')[0])
distinctfilterRDD = onlyIP.distinct()
distinctfilterRDD.collect()
```

```python
outputpath = "output1.txt"
distinctfilterRDD.saveAsTextFile(outputpath)
```

```python
#es 37
inputfile= sc.textFile("sensors.txt")
inputfile.collect()
```

```python
input= inputfile.map(lambda s: (s.split(',')[0], float(s.split(',')[2])))
maxperkey = input.reduceByKey(lambda s1, s2: max(s1, s2))
maxperkey.collect()
```

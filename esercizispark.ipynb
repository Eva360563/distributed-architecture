```python
#es 39bis
inputfile= sc.textFile("sensors.txt")
inputfile.collect()
```

```python
readingsPM1050 = inputfile.filter(lambda s: float(s.split(',')[2]) >=50)
id = inputfile.map(lambda s: (s.split(',')[0], None)).distinct()
interestingreading = readingsPM1050.map(lambda s: (s.split(',')[0], (s.split(',')[1])))
res = interestingreading.groupByKey()
finalSensorCriticalDateStringFormat = res.mapValues(lambda dates : list(dates))
intsensor = finalSensorCriticalDateStringFormat.map(lambda s: (s[0], None))
sub = id.subtract(intsensor)
legati = finalSensorCriticalDateStringFormat.union(sub)

legati.collect()
```

```python
reading= sc.textFile("readings.txt")
reading.collect()
#stationID date hour minute num_of_bikes num_of_free_slot
```

```python
criticalstation= reading.filter(lambda s: int(s.split(',')[5]) <=3)
totsi = reading.map(lambda s: (s.split(',')[0], 1) )
totnumbcritreadingforsi = totsi.reduceByKey(lambda s1,s2: s1 + s2)
si = criticalstation.map(lambda s: (s.split(',')[0], 1) )
numbcritreadingforsi = si.reduceByKey(lambda s1,s2: s1 + s2)
percentage = totnumbcritreadingforsi.join(numbcritreadingforsi)
percentage2 = percentage.mapValues(lambda s:  s[1]/s[0])
percentage2.collect()
```

```python
filtro= percentage2.filter(lambda s: s[1]*100>=80)
sort = filtro.sortBy(lambda s: s[1], False)
sort.collect()
```
